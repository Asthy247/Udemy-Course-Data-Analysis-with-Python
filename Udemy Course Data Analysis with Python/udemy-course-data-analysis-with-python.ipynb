# Bar chart showing Courses per Subject

```python
# Count the number of courses per subject
subject_counts = data['subject'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=subject_counts.index, y=subject_counts.values)
plt.title('Number of Courses per Subject')
plt.xlabel('Subject')
plt.ylabel('Number of Courses')
plt.xticks(rotation=45)
plt.show()
```

**Dominant Subjects:**

Web Development and Business Finance are the most popular subjects on the platform, with a significantly higher number of courses compared to the other two categories.
